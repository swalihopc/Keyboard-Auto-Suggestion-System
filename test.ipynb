```python
import numpy as np
import pandas as pd
import textdistance
import re
from collections import Counter
```

# File Opening And Cleaning 

```python
word = []

with open('story.txt', 'r', encoding='utf-8') as f:
    data = f.read()
    data = data.lower()
    words = re.findall('\w+', data)
    words += words

print(words[0:10])
```

# Make vocabulary

```python
len(words)
```

```python
V = set(words)
V
```

```python
len(V)
```

# Build the Frequency of those words

```python
word_freq_dict=Counter(words)
```

```python
word_freq_dict.most_common(10)
```

# Relative frequency of words

```python
Total_words_freq = sum(word_freq_dict.values())

probs = {}
for k in word_freq_dict.keys():
    probs[k] = word_freq_dict[k] / Total_words_freq
```

```python
probs
```

# Finding similar words

```python
def autocorrect(word):
    word = word.lower()
    if word in V:
        return('Your word seems to be correct', word)
    else:
        similarities = [1-(textdistance.Jaccard(qval=2).distance(v,word))  for v in word_freq_dict.keys()]
        df = pd.DataFrame.from_dict(probs, orient='index').reset_index()
        df = df.rename(columns={'index':'Word',0:'Prob'})
        df['Similarity'] = similarities
        output = df.sort_values(['Similarity','Prob'],ascending=False).head(5)
        return(output)
autocorrect("joi")
```
